```python
import math
import matplotlib.pyplot as plt
import sys
sys.path.append("../lib")
from pylib import pltInit,Point,DataFile
```

```python
plt.rcParams['figure.subplot.bottom']=0.25
plt.rcParams['figure.subplot.left']=0.25
```

```python
def generateData():
    max = 10**6
    k = 8
    xList = list()
    yList = list()
    y = 0
    for i in range(2,max):
        y += math.log(i)
        if i < 8:
            xList.append(i)
            yList.append(y)
        elif i % k ==0:
            xList.append(i)
            yList.append(y)
            k *= 2
    return xList,yList
```

```python
def draw():
    plt.figure(figsize=(10,8),facecolor='white')
    xList,yList = generateData()
    # plt.xscale('log')
    # plt.yscale('log')
    plt.xlim(1,10**6)
    plt.ylim(1,10**7)
    plt.xlabel('$n$')
    plt.ylabel('$\ln(n!)$')
    plt.loglog()
    plt.xticks([10**x for  x in range(7)])
    plt.yticks([10**x for  x in range(8)])
    plt.scatter(xList,yList,label='$\ln(n!)$')
    zList = [x*math.log(x)-x for x in xList]
    plt.plot(xList,zList,label='$n\ln n-n$',c='r')
    plt.legend()
    plt.savefig('factorial.pdf')
```

```python
pltInit()
draw()
```
