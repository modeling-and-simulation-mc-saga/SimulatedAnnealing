```python
import math
import matplotlib.pyplot as plt
import sys
sys.path.append("../lib")
from pylib import pltInit,Point,DataFile
```

```python
def drawSimulation(filename:str):    
    inputFile = filename+'.txt'
    data = DataFile(inputFile)
    plt.figure(figsize=(15,10),facecolor='white')
    plt.xlim(0,200)
    plt.ylim(0,8000)
    plt.xlabel('annealing steps')
    plt.ylabel('$d$')
    xl, yl = data.readData()
    plt.scatter(xl,yl)
    plt.savefig(filename+'.pdf')
```

```python
def draw(filename:str):
    inputFile = filename+'.txt'
    data = DataFile(inputFile)
    plt.figure(figsize=(10,8),facecolor='white')
    plt.xlim(0,50000)
    plt.ylim(0,0.7)
    plt.xlabel('$t$')
    plt.ylabel('$f_i$')
    for k in range(1,5):
        xl, zl = data.readData(y=k)
        plt.plot(xl,zl)
    plt.savefig(filename+'.pdf')
```

```python
pltInit()
#drawSimulation('Simulation')
draw("SimpleMC-10")
```
